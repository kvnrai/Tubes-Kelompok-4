```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np
```

```python
# Load data
file_path = "day.csv"
df = pd.read_csv(file_path)
```

```python
# Convert date column to datetime format
df['dteday'] = pd.to_datetime(df['dteday'])
df['year'] = df['dteday'].dt.year
df['month'] = df['dteday'].dt.month
df['day'] = df['dteday'].dt.day
```

Di bawah menampilkan analisis distribusi

```python
# Distribusi Penyewaan Sepeda
plt.figure(figsize=(12,6))
plt.bar(df['dteday'], df['cnt'], color='skyblue')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan')
plt.title('Distribusi Penyewaan Sepeda')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribusi Penyewaan berdasarkan Cuaca
plt.figure(figsize=(8,5))
sns.barplot(x=df['weathersit'], y=df['cnt'], palette='viridis')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Penyewaan')
plt.title('Distribusi Penyewaan berdasarkan Cuaca')
plt.show()
```

```python
# Rata-rata Penyewaan Sepeda per Musim
season_avg = df.groupby('season')['cnt'].mean()
plt.figure(figsize=(8,5))
season_avg.plot(kind='bar', color='orange')
plt.xlabel('Musim')
plt.ylabel('Rata-rata Penyewaan')
plt.title('Rata-rata Penyewaan Sepeda per Musim')
plt.show()
```

```python

# Tren Penyewaan Sepeda per Hari dalam Sebulan
daily_trend = df.groupby('day')['cnt'].sum()
plt.figure(figsize=(10,5))
plt.plot(daily_trend, marker='o', linestyle='-')
plt.xlabel('Hari dalam Sebulan')
plt.ylabel('Jumlah Penyewaan')
plt.title('Tren Penyewaan Sepeda per Hari dalam Sebulan')
plt.grid()
plt.show()
```

```python
# Korelasi antara Faktor Cuaca dan Penyewaan Sepeda
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['temp'], y=df['cnt'], alpha=0.5)
plt.xlabel('Suhu')
plt.ylabel('Jumlah Penyewaan')
plt.title('Pengaruh Suhu terhadap Penyewaan Sepeda')
plt.show()
```

```python
# Analisis Time Series Penyewaan Sepeda
ts_data = df[['dteday', 'cnt']].set_index('dteday')
ts_data.plot(figsize=(12,6))
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan')
plt.title('Tren Penyewaan Sepeda dari Waktu ke Waktu')
plt.grid()
plt.show()
```

```python
# Analisis Klaster Penyewaan Sepeda
features = df[['temp', 'hum', 'windspeed', 'cnt']]
kmeans = KMeans(n_clusters=3, random_state=42).fit(features)
df['cluster'] = kmeans.labels_
sns.barplot(x=df['cluster'], y=df['cnt'], estimator=np.mean, palette='coolwarm')
plt.xlabel('Cluster')
plt.ylabel('Rata-rata Penyewaan')
plt.title('Analisis Klaster Penyewaan Sepeda')
plt.show()
```

```python
# Analisis Korelasi Cuaca vs. Penggunaan Sepeda
corr_matrix = df[['temp', 'atemp', 'hum', 'windspeed', 'cnt']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Analisis Korelasi Cuaca vs. Penggunaan Sepeda')
plt.show()
```
